```python
#funções utilizadas
def HexToBin(x): # recebe um número hexadecimal e retorna em binário
    if x == '':
        return x
    elif x != '':
        dec = int(x,16) # hex para dec
        Bin = ''
        while dec > 0:
            Bin = str(dec % 2) + Bin
            dec = dec >> 1
        return Bin

def potenciaResultado(x): # retorna o expoente de um número inteiro
    multi = 1
    y = 0
    while multi < x:
        multi = multi * 2
        y = y + 1
    return y

def StrToInt(x): # recebe um número em formato de string(ex:64K), e retorna um número inteiro
    try:
        if x.index('K'):
            y = x.replace('K' , '')
            z = int(y,10)
            a = z * pow(2,10)
            return a
    except ValueError:
        pass
    try:
        if x.index('M'):
            y = x.replace('M' , '')
            z = int(y,10)
            a = z * pow(2,20)
            return a
    except ValueError:
        pass
    try:
        if x.index('G'):
            y = x.replace('G' , '')
            z = int(y,10)
            a = z * pow(2,30)
            return a
    except ValueError:
        pass

def renomeb(x): # retorna o print de um número inteiro em formato de string(ex:2048 = 2K bits)
    if x > pow(2,10):
        x = x / pow(2,10)
        print(x,'K bits')
    elif x > pow(2,20):
        x = x / pow(2,20)
        print(x,'M bits')
    elif x > pow(2,30):
        x = x / pow(2,30)
        print(x,'G bits')
    elif x > pow(2,40):
        x = x / pow(2,40)
        print(x,'T bits')

def renomeB(x): # retorna o print de um número inteiro em formato de string(ex:2048 = 2K bytes)
    if x > pow(2,10):
        x = x / pow(2,10)
        print(x,'K Bytes')
    elif x > pow(2,20):
        x = x / pow(2,20)
        print(x,'M Bytes')
    elif x > pow(2,30):
        x = x / pow(2,30)
        print(x,'G Bytes')
    elif x > pow(2,40):
        x = x / pow(2,40)
        print(x,'T Bytes')
        
# letras nas respostas sempre maiúsculas

# exercícios:
while True:
    R = input('Digite:\n(1) para mapeamento direto,\n(2) para mapeamento associativo,\n(3) para mapeamento associativo por Conjuntos\n(S) para sair.\n:')
    if R == '1':
        I = int(input('''Digite:\n(1) para calcular a quantidade de bits necessáios para a cache,
(2) para calcular o formato de endereço da cache,
(3) para determinar o endereço da linha, bloco ou conjunto desejado.\n:'''))
        if I == 1:
            print('exemplo 5.2')
            CC = input('capacidade da cache em bytes:')
            largLB = int(input('largura do bloco ou da linha em bytes:'))
            CMP = input('capacidade da MP em bytes:')

            CC_ = StrToInt(CC)
            CMP_ = StrToInt(CMP)

            Qlinhas = CC_  / largLB # bits
            Qblocos = CMP_ / largLB # bits
            QblocosPorLinha = Qblocos / Qlinhas # bits
            
            LargTag = potenciaResultado(QblocosPorLinha)
            
            Tt = Qlinhas * LargTag
            Td = CC_ * 8 # byte para bit

            T = Td + Tt
            print('quantidade de bits necessários para determinar a cache:',T,'ou:')
            renomeb(T)

        elif I == 2:
            print('exemplo 5.3')
            CMP = input('capacidade da MP em bytes:')
            Qlinhas = input('quantidade de linhas:')
            LargLB = int(input('largura do bloco ou da linha em bytes:'))

            CMP_ = StrToInt(CMP)
            Qlinhas_ = StrToInt(Qlinhas)

            Qblocos = CMP_ / LargLB
            QblocosPorLinha = Qblocos / Qlinhas_

            LargTag = potenciaResultado(QblocosPorLinha)
            LargLinha = potenciaResultado(Qlinhas_)
            LargCelula = potenciaResultado(LargLB)

            FEND = LargTag + LargLinha + LargCelula
            print('formato do endereço:',FEND,'bits')
            print('formato da tag:',LargTag,'bits')
            print('formato da linha:',LargLinha,'bits')
            print('formato da célula:',LargCelula,'bits')

        elif I == 3:
            print('exemplo 5.4')
            LargLB = int(input('largura do bloco ou da linha em bytes:'))
            CC = input('capacidade da cache em bytes:')
            END = input('endereço desejado em hexadecimal:')

            CC_ = StrToInt(CC)
            TEND = len(END) * 4
            Qlinhas = CC_ / LargLB

            LargCelula = potenciaResultado(LargLB)
            LargLinha = potenciaResultado(Qlinhas)
            LargTag = TEND - LargLinha - LargCelula

            END = HexToBin(END)
            LEND = list(END)
    
            if LEND == '':
                print('endereço não informado')
            elif len(LEND) < TEND:
                while len(LEND) < TEND:
                    LEND.insert(0,0)
                print('endereço total:',*LEND, sep='')
                print('largura e endereço da tag:',LargTag,'->',*LEND[:LargTag], sep='')
                print('largura e endereço da linha desejada:',LargLinha,'->',*LEND[LargTag:LargTag + LargLinha], sep='')
                print('largura e endereço da célula:',LargCelula,'->',*LEND[LargTag + LargLinha:] ,sep='')
            elif len(LEND) == TEND:
                print('endereço total:',*LEND, sep='')
                print('largura e endereço da tag:',LargTag,'->',*LEND[:LargTag], sep='')
                print('largura e endereço da linha desejada:',LargLinha,'->',*LEND[LargTag:LargTag + LargLinha], sep='')
                print('largura e endereço da célula:',LargCelula,'->',*LEND[LargTag + LargLinha:], sep='')

    elif R == '2':
        I = int(input('''Digite:\n(1) para calcular a quantidade de bits necessáios para a cache,
(2) para calcular o formato de endereço da cache,
(3) para determinar o endereço da linha, bloco ou conjunto desejado.\n:'''))
        if I == 1:
            print('exemplo 5.5')
            CC = input('capacidade da cache em bytes:')
            LargLB = int(input('largura da linha ou do bloco em bytes:'))
            CMP = input('capacidade da MP em bytes:')

            CC_ = StrToInt(CC)
            CMP_ = StrToInt(CMP)

            Qlinhas = CC_ / LargLB # bit
            Qblocos = CMP_ / LargLB # bit
            LargTag = potenciaResultado(Qblocos)

            Td = CC_ * 8 # byte para bit
            Tb = Qlinhas * LargTag # bit

            Tc = Td + Tb
            print('quantidade bits necessários para a memória cache:',Tc,'ou:')
            renomeb(Tc)

        elif I == 2:
            print('exemplo 5.6')
            CMP = input('capacidade da MP em bytes:')
            LargLB = int(input('largura do bloco ou da linha em bytes:'))

            CMP_ = StrToInt(CMP)
            Qblocos = CMP_ / LargLB

            campoCelula = potenciaResultado(LargLB)
            campoBloco = potenciaResultado(Qblocos)

            BE = campoCelula + campoBloco
            print('Tamanho total do endereço:', BE,'bits')
            print('largura de endereço do bloco:', campoBloco,'bits')
            print('largura de endereço da célula:', campoCelula,'bits')

        elif I == 3:
            print('exemplo 5.7')
            LargLB = int(input('largura do bloco ou da linha em bytes:'))
            CC = input('capacidade da cache em bytes:')
            END = input('endereço desejado em hexadecimal:')

            CC_ = StrToInt(CC)

            TEND = len(END) * 4
            campoCelula = potenciaResultado(LargLB)

            exp = pow(2,TEND)
            Qblocos = exp / LargLB
            campoBloco = potenciaResultado(Qblocos)

            END = HexToBin(END)
            LEND = list(END)

            if LEND == '':
                print('endereço não informado')
            elif len(LEND) < TEND:
                while len(LEND) < TEND:
                    LEND.insert(0,0)
                print('endereço do bloco desejado:', *LEND[:campoBloco], sep='')
                print('endereço da célula:', *LEND[campoBloco:], sep='')
                print('endereço total:', *LEND, sep='')
            elif len(LEND) == TEND:
                print('endereço do bloco desejado:', *LEND[:campoBloco], sep='')
                print('endereço da célula:', *LEND[campoBloco:], sep='')
                print('endereço total:', *LEND, sep='')

    elif R == '3':
        I = int(input('''Digite:\n(1) para calcular a quantidade de bits necessáios para a cache,
(2) para calcular o formato de endereço da cache,
(3) para determinar o endereço da linha, bloco ou conjunto desejado.\n:'''))
        if I == 1:
            print('exemplo 5.8')
            CC = input('capacidade da cache em bytes:')
            LargLB = int(input('largura da linha ou do bloco em bytes:'))
            conjunto = int(input('linhas por conjunto:'))
            CMP = input('capacidade da MP em bytes:')

            CC_ = StrToInt(CC)
            CMP_ = StrToInt(CMP)

            Td = CC_ * 8 # byte pra bit
            
            Qblocos = CMP_ / LargLB
            Qlinhas = CC_ / LargLB # divisão corta os bytes
            Qconjuntos = Qlinhas / conjunto

            QblocosPorConjuntos = Qblocos / Qconjuntos
            LargTag = potenciaResultado(QblocosPorConjuntos)

            Tt = (LargTag * Qlinhas) # está em bit
            T = Td + Tt
            print('quantidade de bits necessários para determinar a cache:',T,'ou:')
            renomeb(T)
            

        elif I == 2:
            print('exemplo 5.9')
            CMP = input('capacidade da MP em bytes:')
            conjunto = int(input('linhas por conjunto:'))
            CC = input('capacidade da cache em bytes:')
            LargLB = int(input('largura da linha ou do bloco em bytes:'))

            CMP_ = StrToInt(CMP)
            CC_ = StrToInt(CC)

            Qblocos = CMP_ / LargLB
            Qlinhas = CC_ / LargLB # 32K / 16 = 2^11 e não 2^13 
            Qconjuntos = Qlinhas / conjunto # --> 2^11 / 2^2 = 2^9
            QblocosPorConjuntos = Qblocos / Qconjuntos # --> 2^22 / 2^9 = 2^13


            CampoCelula = potenciaResultado(LargLB)
            CampoConjunto = potenciaResultado(Qconjuntos)
            CampoTag = potenciaResultado(QblocosPorConjuntos)

            BE = CampoCelula + CampoConjunto + CampoTag
            print('formato do barramento de endereço:',BE,'bits')
            print('formato da tag:',CampoTag,'bits')
            print('formato do conjunto',CampoConjunto,'bits')
            print('formato da célula',CampoCelula,'bits')

        elif I == 3:
            print('exemplo 5.10')
            LargLB = int(input('largura da linha ou do bloco em bytes:'))
            CC = input('capacidade da cache em bytes:')
            conjunto = int(input('linhas por conjunto:'))
            END = input('endereço desejado em hexadecimal:')

            CC_ = StrToInt(CC)

            TEND = len(END) * 4
            campoCelula = potenciaResultado(LargLB)
            
            Qlinhas =  CC_ / LargLB 
            Qconjuntos = Qlinhas / conjunto

            campoConjunto = potenciaResultado(Qconjuntos)
            campoCelula = potenciaResultado(LargLB)
            campoTag = TEND - (campoConjunto + campoCelula)

            END = HexToBin(END)
            LEND = list(END)

            if LEND == '':
                print('endereço não informado')
            elif len(LEND) < TEND:
                while len(LEND) < TEND:
                    LEND.insert(0,0)
                print('endereço da tag:', *LEND[:campoTag], sep='')
                print('endereço do conjunto desejado:', *LEND[campoTag:campoTag + campoConjunto], sep='')
                print('endereço da célula:', *LEND[campoTag + campoConjunto:], sep='')
                print('endereço total:', *LEND, sep='')
    elif R == 'S':
        break
```
